```python
import openai
import config
openai.api_key = config.API_KEY
```

```python
question = input("What would you like to ask ChatGPT?") # how to swim?
```

```python
# a json object is returned
response = openai.chat.completions.create(
    model = "gpt-4", # varient/version of GPT model
    messages = [
            {"role": "user", "content": f"{question}"},
        ],
    max_tokens = 1024, # max num of tokens the request and response should contain together
    n=1, # num of completions to generate for the prompt
    stop = None, # the sequence where the API should stop generating the response
    temperature = 0.8, # controls creativity of generated response: 1-->most creative but less coherent; 0 --> more predictable but less interesting
)
```

- choices: array of objects with generated responses; if n=1, only 1 response object generated
- text: the actual response generated by the GPT-3 model
- finish_reason: reason why the response was generated (here because model reached the stop condition provided in the request. stop=None--> full response from the ChatGPT API was returned)
- created: specifies Unix timestamp of when the response was created
- id: unique identifier for the API request that generated this response
- model: the GPT-3 model used to generate the response
- object: the type of object that was returned (here it is text_completion object)
- usage: info about resource usage of the API request (num of tokens used for completion; num of tokens in the prompt; total num of tokens used)


### Most important param from response: text 


```python
answer = "".join(response.choices[0].message.content.strip())
print(answer)
```
